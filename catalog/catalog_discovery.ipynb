# Remote Catalog Discovery (STAC/OData)**Source**: `grdl/example/catalog/catalog_discovery.py`## Learning Objectives- Query remote catalogs (BIOMASS STAC, Sentinel-1/2 OData, NISAR Earthdata)- Filter by spatial extent and time range- Browse available products## Theory: Catalog Protocols**STAC** (SpatioTemporal Asset Catalog): JSON-based metadata standard  **OData**: Copernicus Data Space Ecosystem query API  **Earthdata**: NASA's CMR (Common Metadata Repository)## Data Setup**Authentication**: Some catalogs require credentials. Configure via:```bash# Copernicus Data Space (Sentinel-1/2)export CDSE_USERNAME=your_usernameexport CDSE_PASSWORD=your_password# NASA Earthdata (NISAR, ASTER, VIIRS)export EARTHDATA_USERNAME=your_usernameexport EARTHDATA_PASSWORD=your_password```Or save credentials to `~/.config/geoint/credentials.json`.
